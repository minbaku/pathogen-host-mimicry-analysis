```python
# 01_peptide_similarity.ipynb
!pip install biopython --quiet
from Bio import SeqIO  # fixed capitalisation

fasta_path_pathogen = '/content/Pathogen.txt'   # fixed to relative path
fasta_path_human = '/content/Human.txt'          # fixed to relative path

def load_single_sequence(fasta_path):
    for record in SeqIO.parse(fasta_path, "fasta"):
        return str(record.seq)

def generate_peptides(sequences, k=9):
    return [sequences[i:i+k] for i in range(len(sequences) - k + 1)]
    # fixed space after return

pathogen_sequence = load_single_sequence(fasta_path_pathogen)
human_sequence = load_single_sequence(fasta_path_human)

pathogen_peptides = generate_peptides(pathogen_sequence, k=9)
human_peptides = generate_peptides(human_sequence, k=9)

def exact_matches(x, y):
    return set(x).intersection(set(y))

matches = exact_matches(pathogen_peptides, human_peptides)

print(matches)
print("Number of matches:", len(matches))
```
